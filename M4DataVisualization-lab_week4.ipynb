## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
import matplotlib.pyplot as plt
"""
plt.scatter(x = df['Age'],y = df['WorkWeekHrs'],marker='+')
plt.title('Scatter plot of Age and WorkWeekhrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.show()
"""
#we can create a group by age with avg workweekhrs
df_group = df[['Age','WorkWeekHrs']]
df_age_groupby = df_group.groupby(['Age'],as_index = False).mean()
#df_age_groupby
#we can also create using sqllite query
#Query = '''select Age,Avg(WorkWeekHrs) from master group by age'''
#df_age_groupby = pd.read_sql_query(Query,conn)
df_age_groupby.plot(kind = 'scatter',x ='Age',y='WorkWeekHrs',marker = '+',color ='red')
plt.title('Scatter plot of Age vs avg WorkWeekhrs')
plt.xlabel('Age')
plt.ylabel('avg WorkWeekHrs')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
Query2 ="""
select Age, WorkWeekHrs, CodeRevHrs 
from master
"""
df_bub =pd.read_sql_query(Query2,conn)
plt.scatter(x='WorkWeekHrs',y='CodeRevHrs', s='Age',alpha =0.5, data =df_bub)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
Query4 ="""
Select count(DatabaseDesireNextYear) as count,DatabaseDesireNextYear from DatabaseDesireNextYear
group by DatabaseDesireNextYear
order by count desc limit 5
"""
df2= pd.read_sql_query(Query4,conn)
label =['PostgreSQL','MongoDB','Redis','MySQL','Elasticsearch']
plt.pie(df2['count'],labels =label,autopct='%1.1f%%')
plt.title('Top5 databases')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
df_30_35 = df[(df['Age']>=30) &(df['Age']<=35)]
df_30_35 = df_30_35[['WorkWeekHrs','CodeRevHrs','Age']]
df_30_35_age = df_30_35.groupby(['Age']).median()
df_30_35_age
```

```python
labels = [30,31,32,33,34,35]
width = 0.35      

fig, ax = plt.subplots()

ax.bar(labels, df_30_35_age['WorkWeekHrs'], width, label='WorkWeekHrs')
ax.bar(labels, df_30_35_age['CodeRevHrs'], width, label='CodeRevHrs')

ax.set_ylabel('WorkWeekHrs and CodeRevHrs')
ax.set_title('Median WorkWeekHrs and CodeRevHrs for the age group 30 to 35')
ax.legend()
```

Plot the median `ConvertedComp` for all ages from 45 to 60.
